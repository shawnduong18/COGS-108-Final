```python
# Reformat Column Titles
def fix_column_labels(column_name):
    column_name = column_name.strip()
    column_name = column_name.replace('\n', ' ')
    column_name = column_name.upper()
    return column_name

df.columns = df.columns.map(fix_column_labels)
df.head()
```

```python
# Sorting DataFrame by offensive and defensive stats
df_offense = df[['TEAM','GP','W','L','YEAR', 'PTS','AST', 'TOV', 'FT%', '3P%', 'FG%']]
df_defensive = df[['TEAM','GP','W','L', 'YEAR', 'DREB', 'OPP 3P%', 'OPP PTS PAINT', 'OPP FG%', 'OPP TOV', 'OPP PTS']]

# doing a bit of cleaning
df_offense['W%'] =  df_offense['W']/df_offense['GP']
df_defensive['W%'] = df_defensive['W']/df_defensive['GP']


df_offense = df_offense.drop(columns = ['GP','W','L'])
df_defensive = df_defensive.drop(columns = ['GP','W','L'])

# fixing inconsistency
df_offense['TEAM'] = df_offense['TEAM'].replace('Los Angeles Clippers', 'LA Clippers')
df_defensive['TEAM'] = df_defensive['TEAM'].replace('Los Angeles Clippers', 'LA Clippers')
```

```python
# we will start by extracting each individual season from our dataframe
df_defensive_14_15 = df_defensive[df_defensive['YEAR'] == '2014-2015']
df_defensive_15_16 = df_defensive[df_defensive['YEAR'] == '2015-2016']
df_defensive_16_17 = df_defensive[df_defensive['YEAR'] == '2016-2017']
df_defensive_17_18 = df_defensive[df_defensive['YEAR'] == '2017-2018']
df_defensive_18_19 = df_defensive[df_defensive['YEAR'] == '2018-2019']
df_defensive_19_20 = df_defensive[df_defensive['YEAR'] == '2019-2020']
df_defensive_20_21 = df_defensive[df_defensive['YEAR'] == '2020-2021']
df_defensive_21_22 = df_defensive[df_defensive['YEAR'] == '2021-2022']
df_defensive_22_23 = df_defensive[df_defensive['YEAR'] == '2022-2023']
df_defensive_23_24 = df_defensive[df_defensive['YEAR'] == '2023-2024']

# will be useful for populating the graphs
datasets = [
    (df_defensive_14_15, "2014-2015"),
    (df_defensive_15_16, "2015-2016"),
    (df_defensive_16_17, "2016-2017"),
    (df_defensive_17_18, "2017-2018"),
    (df_defensive_18_19, "2018-2019"),
    (df_defensive_19_20, "2019-2020"),
    (df_defensive_20_21, "2020-2021"),
    (df_defensive_21_22, "2021-2022"),
    (df_defensive_22_23, "2022-2023"),
    (df_defensive_23_24, "2023-2024")
]

# creating dictionary that will store all the rankings we need
rankings_def = {}
```

```python
# now we can begin graphing each stat per year
# this will look at OPPONENT FIELD GOAL PERCENTAGE
fig, axes = plt.subplots(nrows = 2, ncols = 5, figsize = (20,10))
fig.suptitle('Opponent Field Goal Percentage')

# taking an automated approach to filling the graph
# looping through datasets to generate plots
for i, (df,year) in enumerate(datasets):
    # getting the graphs to wrap around after reaching the end
    row, col = divmod(i,5)

    # sorting the values as it will be useful for later analysis
    sorted = df.sort_values(by = "OPP FG%", ascending = True)

    # creating a rank series to store it later
    sorted['RANK'] = range(1, len(sorted) + 1)

    # making sure the year appears in the dictionary so then we can add information to it
    if year not in rankings_def:
        rankings_def[year] = {}
        
    # storing the ranks
    rankings_def[year]['OPP FG%'] = sorted[['TEAM', 'OPP FG%', 'RANK']]

    # plotting the data
    sns.barplot(data = sorted, y = 'TEAM', x = 'OPP FG%', ax = axes[row,col], color = 'blue')

    # setting the title
    axes[row,col].set_title(year)

# clearing the team names on the side for better legibility
for i in range(0,5):
    axes[0,i].tick_params(axis = 'y', labelsize = 10)
    axes[1,i].tick_params(axis = 'y', labelsize = 10)

# getting rid of the overlap
plt.tight_layout(rect = [0, 0, 1, 1])

# getting rid of redundant y-label
for i in range(1,5):
    axes[0,i].set_ylabel('')
    axes[1,i].set_ylabel('')
```

As we see here, we have a clear ranking for each indivudal season for the last 10 years of average opponent field goal percentage. The data is sorted in ascending order, as a **good defense** should allow a **lower field goal percentage** to their opponents, essentially meaning the average opponent does not score a good amount of field goals compared to how much they take.

```python
# this will look at OPPONENT TURNOVER RATE
fig, axes = plt.subplots(nrows = 2, ncols = 5, figsize = (20,10))
fig.suptitle('Opponent Turnover Rate')

# looping through datasets to generate plots
for i, (df,year) in enumerate(datasets):
    # getting the graphs to wrap around after reaching the end
    row, col = divmod(i,5)

    # sorting the values as it will be useful for later analysis
    sorted = df.sort_values(by = "OPP TOV", ascending = False)

    # storing the ranks as it will be useful later
    sorted['RANK'] = range(1, len(sorted) + 1)
    rankings_def[year]['OPP TOV'] = sorted[['TEAM', 'OPP TOV', 'RANK']]

    # plotting the data
    sns.barplot(data = sorted, y = 'TEAM', x = 'OPP TOV', ax = axes[row,col], color = 'blue')

    # setting the title
    axes[row,col].set_title(year)

# clearing the team names on the side for better legibility
for i in range(0,5):
    axes[0,i].tick_params(axis = 'y', labelsize = 10)
    axes[1,i].tick_params(axis = 'y', labelsize = 10)

# getting rid of the overlap
plt.tight_layout(rect = [0, 0, 1, 1])

# getting rid of redundant y-label
for i in range(1,5):
    axes[0,i].set_ylabel('')
    axes[1,i].set_ylabel('')
     
```

# Getting Our Answer

Here we will use the rankings we collected to calculate our defensive and offensive ratings to then compare to regular season win percentage. The idea is to utilize the rankings we have collected to then gather a team's overall rankings across all 6 statistics we measure per each side of the ball. In other words, let's say a team ranked 1st across all 6 statistics on the defensive side of the ball, we would take their average (which in this case would be 1) along with every other team's averages. Then, we will plot each team's average and merge all the data over the 10 years to then perform regression and using the data to eventually come to a conclusion

So here, we are calculating the averages for each team and storing in a separate dictionary that will then be used to create plots

```python
# dictionary that will store a teams offensive rankings per season
offensive_rankings = {}

# iterating through our dictionary of rankings that contain a year and a defensive stat
for year, stats_dict in ranking_off.items():
    # creating a temporary dictionary to store new values we will calculate
    team_ranks = {}

    # iterating through the individual stat
    for stat,df in stats_dict.items():
        # each row stores team, the actual value of the stat, and its corresponding rank in comparison to the rest of the league that year
        for _, row in df.iterrows():
            team = row['TEAM']
            rank = row['RANK']

            # checking if the team is not already put into ranks to add it
            if team not in team_ranks:
                team_ranks[team] = []
            # adding the respective rank of the team
            team_ranks[team].append(rank)

    # now that we are out of the individual stats, we know we have collected the rank for each of our 6 defensive stats
    # we will calculate their average rank
    avg_ranks = {team: sum(ranks) / len(ranks) for team, ranks in team_ranks.items()}

    # creating a dataframe that stores the final average rank per team
    final_rank = pd.DataFrame(avg_ranks.items(), columns=['TEAM','AVG_RANK']).sort_values(by='AVG_RANK')

    # creating a series corresponding to final rank
    final_rank['FINAL_RANK'] = range(1, len(final_rank) + 1)

    # adding it to the dictionary
    offensive_rankings[year] = final_rank
```

Now, we are storing ranks along with win percentage, to make plotting the two easier.

```python
# storing a team's defensive ranks plus their win percentage for the season
offense_and_wins = {}

# iterating through the year and rank
for year, def_rank in offensive_rankings.items():
    # extracting the win percentage
    win_percentage = df_offense[df_offense['YEAR'] == year][['TEAM', 'W%']]

    # merging the two to add win percentage to the library
    combined = pd.merge(def_rank, win_percentage, on='TEAM')

    # adding to a new dictionary
    offense_and_wins[year] = combined
```

## Graphing Offensive Rank vs Win Percentage
Here, we will begin to create our scatterplots to see if we find any sort of correlation between a team's offensive ranking and their regular season win percentage. Once we take a look at the graphs to gauge what we're looking for, we will perform regression.

```python
# grouping the plots to save space
fig, axes = plt.subplots(nrows=2, ncols=5, figsize=(20,10))

axes = axes.flatten()

# storing the legend separately so it doesn't take up space on other graphs
team_colors = {}

# scatterplot
for i, (year, df) in enumerate(offense_and_wins.items()):
    ax = axes[i]

    # creating scatterplot
    scatter = sns.scatterplot(x=df['AVG_RANK'], y=df['W%'], ax=ax, hue=df['TEAM'],legend=False)

    sns.regplot(x=df['AVG_RANK'], y=df['W%'], ax=ax, scatter=False, color="black", line_kws={"linewidth": 2})

    # GOT IDEA TO VISUALIZE LEGEND SEPARATELY FROM CHATGPT
    if not team_colors:
        team_colors = {t: c for t, c in zip(df['TEAM'], scatter.get_children()[0].get_facecolors())}

    # making scatterplots more readable
    ax.invert_xaxis()
    ax.set_title(f"{year}")
    ax.set_xlabel("Average Offensive Rank")
    ax.set_ylabel("Win Percentage")

# graphing scatterplots
plt.tight_layout()
plt.show()

# creating this for legend plot
# IDEAS FROM CHATGPT
fig_legend = plt.figure(figsize=(8,6))
ax_legend = fig_legend.add_subplot(111)
ax_legend.axis("off")

handles = [plt.Line2D([0],[0], marker='o', color='w', markerfacecolor=color,markersize=10)
           for color in team_colors.values()]
ax_legend.legend(handles, team_colors.keys(), title="Teams",loc="center",frameon=False)
plt.show()
```

```python
combined_years = pd.concat(offense_and_wins.values(), ignore_index=True)

combined_years = combined_years.rename(columns={'W%': 'Win_Percentage'})
outcome, predictors = patsy.dmatrices('Win_Percentage ~ FINAL_RANK', combined_years)
model = sm.OLS(outcome, predictors).fit()

print(model.summary())
```
